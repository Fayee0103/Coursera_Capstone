```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate
```

```python
res = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))
df2=df[0]
print( tabulate(df2, headers='keys', tablefmt='psql') )
```

```python
# define the dataframe columns
column_names = ['FSA','Borough', 'Neighborhood'] 

# instantiate the dataframe
df2.columns=column_names
df2
```

```python
# delete all rows with condition  
indexNames = df2[ (df2['Borough'] =='Not assigned' ) & (df2['Neighborhood'] =='Not assigned') ].index
df2.drop(indexNames,inplace=True)
```

```python
df2.reset_index(drop=True, inplace=True)
df2
```

```python
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])
```

```python
df2
```

```python
df3=df2.groupby(['FSA','Borough']).agg(lambda x: ', '.join(set(x))).reset_index()
df3
```
